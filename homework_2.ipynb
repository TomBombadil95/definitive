#### H0: The null hypothesis states that there is no correlation between the side on which a team plays and the result of the match

#### H1: The alternative hypothesis states that there exists correlation between the side on which a team plays and the result of the match

### Chi-Square test

```python
import numpy as np
f_obvs=np.append(cont_tab.iloc[0][0:3].values, cont_tab.iloc[1][0:3].values)
f_obvs
```

```python
row_sum=cont_tab.iloc[0:2,3].values
row_sum
```

```python
col_sum=cont_tab.iloc[2,0:3].values
col_sum
```

```python
total=cont_tab.loc['All','All']
f_expected=[]
for j in range(2):
    for i in col_sum:
        f_expected.append(i*row_sum[j]/total)
f_expected
```

```python
chi_square_stat=((f_obvs-f_expected)**2/f_expected).sum()
print('Chi_square_stat: {}'.format(chi_square_stat))
```

```python
dof=(len(row_sum)-1)*(len(col_sum)-1)
print("Degrees of freedom: {}".format(dof))
```

```python
import scipy.stats as ss
```

```python
ss.chi2.ppf(0.95,2)<chi_square_stat
```

#### The value of the test statistics of Chi-square test obtained in our calculation is greater than 0.95 percentile of the chi-square (with 2 degrees of freedom), we can conclude that the null hypothesis is rejected. There is a relationship between the side on which a team plays and it's result in a match. Moreover, we can conclude from the bar-plot that playing home matches influences positively a team's chance of winning the match.

## Now I'm able to see all the passes and the correspondent player ID

```python
passes=newengevents
passes=passes.drop('positions',axis=1)
passes.playerId=list(passes.playerId)
PLAYERS=[]
for i in range(len(passes.playerId)):
    PLAYERS.append(passes.playerId.iloc[i])
passes_player=pd.DataFrame(PLAYERS , columns=['PlayerId'])
passes_player['Numberofpasses'] = passes_player.groupby('PlayerId')['PlayerId'].transform('count')
passes_player
```

## Now I eliminate that players that don't have the minimum number of passes
In these lines we compute the median and the standard deviation of the frequency of passes. 
We count out the player that doesn't have a sufficient number of passes considering $threshold_{passes} = medium_{passes} - std_{passes}$. We chose this value as the threshold value as we considered that the significant values are the ones that are greater than the medium value minus his lower dispersion.

```python
Median= passes_player.Numberofpasses.median()
STD=passes_player.Numberofpasses.std()
Median=int(Median)
STD=int(STD)
T=Median-STD
ID_passes_threshold=[]
frequency_passes_threshold=[]
frequency_passes=[]
ID_passes=[]
for i in range(len(passes_player.Numberofpasses)):
    frequency_passes.append(float(passes_player.Numberofpasses.iloc[i]))
    ID_passes.append(float(passes_player.PlayerId.iloc[i]))
for i in range(len(frequency_passes)):
    if frequency_passes[i]>T:
        frequency_passes_threshold.append(int(frequency_passes[i]))
        ID_passes_threshold.append(int(ID_passes[i]))
```

```python
passes_threshold=pd.DataFrame(ID_passes_threshold, columns=['playerId'])
passes_threshold.insert(1, "Numberofpasses",frequency_passes_threshold, True)
passes_threshold.drop_duplicates(subset='playerId', keep='first', inplace=True)
passes_threshold
```

```python
#Design the football pitch
from matplotlib.patches import Rectangle
from matplotlib.patches import ConnectionPatch
from matplotlib.patches import Arc
def draw_pitch(ax):
    # focus on only half of the pitch
    #Pitch Outline & Centre Line
    Pitch = Rectangle([0,0], width = 120, height = 80, fill = False)
    #Left, Right Penalty Area and midline
    LeftPenalty = Rectangle([0,22.3], width = 14.6, height = 35.3, fill = False)
    RightPenalty = Rectangle([105.4,22.3], width = 14.6, height = 35.3, fill = False)
    midline = ConnectionPatch([60,0], [60,80], "data", "data")

    #Left, Right 6-yard Box
    LeftSixYard = Rectangle([0,32], width = 4.9, height = 16, fill = False)
    RightSixYard = Rectangle([115.1,32], width = 4.9, height = 16, fill = False)


    #Prepare Circles
    centreCircle = plt.Circle((60,40),8.1,color="black", fill = False)
    centreSpot = plt.Circle((60,40),0.71,color="black")
    #Penalty spots and Arcs around penalty boxes
    leftPenSpot = plt.Circle((9.7,40),0.71,color="black")
    rightPenSpot = plt.Circle((110.3,40),0.71,color="black")
    leftArc = Arc((9.7,40),height=16.2,width=16.2,angle=0,theta1=310,theta2=50,color="black")
    rightArc = Arc((110.3,40),height=16.2,width=16.2,angle=0,theta1=130,theta2=230,color="black")
    
    element = [Pitch, LeftPenalty, RightPenalty, midline, LeftSixYard, RightSixYard, centreCircle, 
               centreSpot, rightPenSpot, leftPenSpot, leftArc, rightArc]
    for i in element:
        ax.add_patch(i)
```

```python
pitch=plt.figure() #set up the figures
pitch.set_size_inches(7, 5)
axes=pitch.add_subplot(1,1,1)
draw_pitch(axes) #overlay our different objects on the pitch
plt.ylim(-2, 82)
plt.xlim(-2, 122)
plt.axis('off')
plt.show()
```
